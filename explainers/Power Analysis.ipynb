```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats
```

# Power of one-sample t-test

```python
x = np.linspace(-3, 6, 200)
mu0 = 0
s = 1
df = 10
t = stats.t(df=df, loc=mu0, scale=s)
pdf_h0 = t.pdf(x)
t_crit = t.ppf(.975)

fig, ax = plt.subplots(figsize=(6.4, 3.8), dpi=120)

def plot_H0(ax):
    ax.hlines(0, -3, 6, color='black')
    ax.text(-2, .3, r'$P(t|H_0)$', ha='center', c='C0')
    ax.plot(x, pdf_h0, label='$P(t|H_0)$')
    ax.fill_between(x, pdf_h0, where=np.logical_or(x > t_crit, x < -t_crit), color='C0', alpha=.3)
    ax.text(2.3, .015, r'$\alpha/2$', ha='center', c='C0')
    ax.text(-2.3, .015, r'$\alpha/2$', ha='center', c='C0')
    ax.vlines(mu0, .01, -.01)
    ax.text(mu0, -.05, r'$\mu_0$', ha='center', c='C0')
    
    ax.text(t_crit, -.05, r'${t_{crit}}$', ha='center')
    ax.axis('off');
    
plot_H0(ax)
```

# Lachin 1981

```python
x = np.linspace(-3, 6, 200)
mu0 = 0
sigma = 1
pdf_h0 = stats.norm.pdf(x, mu0, sigma)
X_alpha = stats.norm.ppf(.95, mu0, sigma)
mu1 = X_alpha + stats.norm.ppf(.8)
pdf_h1 = stats.norm.pdf(x, mu1, sigma)

plt.axhline(0, c='black')

plt.text(-2, .3, r'$P(X|H_0)$', ha='center', c='C0')
plt.plot(x, pdf_h0, label='$P(X|H_0)$')
plt.fill_between(x, pdf_h0, where=x>X_alpha, color='C0', alpha=.3)
plt.text(2, .02, r'$\alpha$', ha='center', c='C0')
plt.axvline(mu0, alpha=0.5)
plt.text(mu0, -.05, r'$\mu_0$', ha='center', c='C0')

plt.text(4.5, .3, r'$P(X|H_1)$', ha='center', c='C1')
plt.plot(x, pdf_h1, label='$P(X|H_0)$')
plt.fill_between(x, pdf_h1, where=x<X_alpha, color='C1', alpha=.3)
plt.text(1, .05, r'$\beta$', ha='center', c='C1')
plt.axvline(mu1, alpha=0.5, c='C1')
plt.text(mu1, -.05, r'${\mu_1}$', ha='center', c='C1')

plt.axvline(X_alpha, alpha=0.5, c='gray')
plt.text(X_alpha, -.05, r'${X_\alpha}$', ha='center')

plt.gca().axis('off');
```
