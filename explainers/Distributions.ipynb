```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
import sklearn
from scipy import stats
```

```python
ff = [np.random.normal, np.random.uniform, np.random.lognormal]
for rs in ff:
    ll = []
    pp = np.zeros(100)
    for n in range(3, 51, 5):
        for ii in range(100):
            X = rs(size=n)
            _, pp[ii] = stats.shapiro(X)
        ll.append((n, (pp>0.05).sum()/100))
    arr = np.array(ll)
    plt.plot(arr[:,0], arr[:,1], '-', label=str(rs).split()[2])
# plt.ylim((0,1))
plt.legend()
plt.gca().set(xlabel='sample size', ylabel='normal not rejected', title='Shapiro-Wilk test power')
```
